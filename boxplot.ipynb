```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('coursework.csv')
```

```python
m_r = data['all_male']
m_w = data['all_female']
```

```python
m_r
```

```python
output_filename1='all_man_ratio.png'

figure_width, figure_height = 4,10
fig=plt.figure(figsize=(figure_width,figure_height))
plt.xlabel('All men',fontsize=20)
plt.ylabel('Obesity_ratio',fontsize=20)
plt.xlim([0.75,1.25])
plt.xticks([])
plt.boxplot(m_r,manage_ticks=(False))
plt.savefig(output_filename1,bbox_inches='tight')

output_filename2='all_women_ratio.png'
figure_width, figure_height = 4,10
fig=plt.figure(figsize=(figure_width,figure_height))
plt.xlabel('All women',fontsize=20)
plt.ylabel('Obesity_ratio',fontsize=20)
plt.xlim([0.75,1.25])
plt.xticks([])
plt.boxplot(m_w,manage_ticks=(False))
plt.savefig(output_filename2,bbox_inches='tight')
```

```python
data1={
    'ALL men':list(m_r),
    'All women':list(m_w)
}
figure_width, figure_height = 15,10
plt.figure(figsize=(figure_width,figure_height))
box_1,box_2=data1['ALL men'],data1['All women']
#plt.subplot(222)
labels='ALL men','All women'
#plt.xlabel('people',fontsize=20)
plt.ylabel('Obesity_ratio(%)',fontsize=20)
#plt.xlim([0.75,1.25])
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.boxplot([box_1,box_2],notch=False, labels = labels)
plt.savefig('male vs female',dpi=200,bbox_inches='tight')
```

```python
l_m=data['l_male']
l_f=data['l_female']
m_m=data['mb_male']
m_f=data['mb_female']
n_m=data['nm_male']
n_f=data['nm_female']
u_m=data['ua_male']
u_f=data['ua_female']
data2={
    'l_male':list(l_m),
    'l_female':list(l_f),
    'mb_male':list(m_m),
    'mb_female':list(m_f),
    'nm_male':list(n_m),
    'nm_female':list(n_f),
    'ua_male':list(u_m),
    'ua_female':list(u_f)
}
```

```python
l_m=[l_m_ for l_m_ in l_m if l_m_ == l_m_ ]
l_f=[l_f_ for l_f_ in l_f if l_f_ == l_f_ ]
m_m=[m_m_ for m_m_ in m_m if m_m_ == m_m_ ]
m_f=[m_f_ for m_f_ in m_f if m_f_ == m_f_ ]
n_m=[n_m_ for n_m_ in n_m if n_m_ == n_m_ ]
n_f=[n_f_ for n_f_ in n_f if n_f_ == n_f_ ]
u_m=[u_m_ for u_m_ in u_m if u_m_ == u_m_ ]
u_f=[u_f_ for u_f_ in u_f if u_f_ == u_f_ ]
```

```python
figure_width, figure_height = 15,10
plt.figure(figsize=(figure_width,figure_height))
box_1,box_2,box_3,box_4,box_5,box_6,box_7,box_8=l_m,l_f,m_m,m_f,n_m,n_f,u_m,u_f
labels='l_male','l_female','mb_male','mb_female','nm_male','nm_female','ua_male','ua_female'
plt.ylabel('Obesity_ratio(%)',fontsize=20)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.boxplot([box_1,box_2,box_3,box_4,box_5,box_6,box_7,box_8],notch=False, labels = labels)
plt.savefig('male vs female in different authority',dpi=300,bbox_inches='tight')
```
